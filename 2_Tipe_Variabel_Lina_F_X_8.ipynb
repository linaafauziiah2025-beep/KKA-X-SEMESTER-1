```python
#Pembagian bulat

total_botol_air_minum = 30
jumlah_anak_yatim = 7

jumlah_botol_tiap_anak_yatim = total_botol_air_minum // jumlah_anak_yatim

print(jumlah_botol_tiap_anak_yatim)
```
